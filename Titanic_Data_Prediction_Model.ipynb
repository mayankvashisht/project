```python
train_df = df_train[["PassengerId","Survived","Pclass","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"]]
test_df = df_test[["PassengerId","Pclass","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"]]
```

```python
train_df.isnull().sum()
```

```python
train_df.drop("Cabin", inplace = True, axis = 1)
test_df.drop("Cabin", inplace = True, axis = 1)
train_df
```

```python
train_df.drop("Ticket", inplace = True, axis = 1)
test_df.drop("Ticket", inplace = True, axis = 1)
train_df
```

```python
train_df.groupby("Embarked").size()
```

```python
def function(x):
    if (x["Embarked"] == "C"):
        return int(1)
    elif(x["Embarked"] == "Q"):
        return int(2)
    elif(x["Embarked"] == "S"):
        return int(3)
train_df1 = train_df.apply(function, axis = 1)
test_df1 = test_df.apply(function,axis = 1)
```

```python
train_df["Embarked"] = train_df1
test_df["Embarked"] = test_df1
```

```python
train_df.head()
```

```python
train_df["Age"] = train_df["Age"].fillna(train_df["Age"].mean())
test_df["Age"] = test_df["Age"].fillna(test_df["Age"].mean())
train_df["Age"] = train_df["Age"].astype(int)
train_df.shape
```

```python
test_df.head()
```

```python
test_df.shape
```

```python
test_df.isnull().sum()
```

```python
train_df["Embarked"] = train_df["Embarked"].fillna(train_df["Embarked"].mean())
test_df["Fare"] = test_df["Fare"].fillna(test_df["Fare"].mean())
```

```python
train_df.info()
test_df.info()
```

```python
train_df.describe()
```

```python
def function(x):
    if (x["Sex"] == "male"):
        return int(0)
    elif(x["Sex"] == "female"):
        return int(1)
   
train_df2 = train_df.apply(function, axis = 1)
test_df2 = test_df.apply(function,axis = 1)
train_df["Sex"] = train_df2
test_df["Sex"] = test_df2
train_df.head()
```

```python
test_df.head()
```

```python
x = train_df[["PassengerId","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
y = train_df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
x_train.info()
```

```python
def get_mae(max_leaf_nodes, x_train, x_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes= max_leaf_nodes , random_state=0)
    model.fit(x_train, y_train,)
    preds_val = model.predict(x_test)
    mae = mean_absolute_error(y_test, preds_val)
    r_square = (round(r2_score(y_test,preds_val)*100, 2))
    return mae, r_square
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae, r_square = get_mae(max_leaf_nodes, x_train, x_test, y_train, y_test)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d \t\t R-squared:  %d" %(max_leaf_nodes, my_mae, r_square))
```

```python
model = DecisionTreeRegressor(max_leaf_nodes= 5 , random_state=0)
model.fit(x_train, y_train,)
preds_val = model.predict(x_test)
mae = mean_absolute_error(y_test, preds_val)
print("mae = ",mae, "accuracy score = " , model.score(x_test,y_test)*100)
```

```python
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
predict = linear.predict(x_test)
mae = mean_absolute_error(y_test, predict)
r_squared = (round(r2_score(y_test, predict)*100, 2))
print("mean squared error: ", mae, "r_square: ", r_squared )
print("accuracy score = ", linear.score(x_test, y_test)*100)
```

```python
logistic = LogisticRegression(max_iter = 10000)
logistic.fit(x_train,y_train)
```

```python
predicted = logistic.predict(x_test)
mae = mean_absolute_error(y_test, predicted)
r_squared = (round(r2_score(y_test, predicted)*100, 2))
print("mean squared error: ", mae, "r_square: ", r_squared )
print("accuracy score = ", logistic.score(x_test, y_test)*100)
```

```python
forest_model = RandomForestRegressor(random_state=16)
forest_model.fit(x_train, y_train)
preds = forest_model.predict(x_test)
mae = mean_absolute_error(y_test, preds)
r_squared = (round(r2_score(y_test, preds)*100, 2))
print("mean squared error: ", mae, "r_square: ", r_squared )
print("accuracy score = ", forest_model.score(x_test, y_test)*100)
```

```python
model2= GaussianNB(var_smoothing=1e-08)
model2.fit(x_train, y_train)
Y_pred= model2.predict(x_test)
model2.score(x_test, y_test)
```
